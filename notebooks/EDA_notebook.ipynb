# 3. What should the discount rate be good for the profit margin?

```python
#creat dataframe with sum of sales BEFORE discount
sales_org = df.groupby(['Product Name'],as_index=False)['Sales'].sum()
sales_org.columns=['Product Name','Original sales']

#creat dataframe with sum of sales AFTER discount
sales_dis = df.groupby(['Product Name'],as_index=False)['Order Item Total'].sum()
sales_dis.columns=['Product Name','Discount sales']

#creat dataframe with sum of profit AFTER discount
pf_dis = df.groupby(['Product Name'],as_index=False)['Order Profit Per Order'].sum()
pf_dis.columns=['Product Name','Discount profit']
```

```python
combine_1 = sales_org.merge (sales_dis, on='Product Name')
combine_3 = combine_1.merge (pf_dis, on='Product Name')

# Calculate margin AFTER discount
# Fomula: margin = profit/revenue
combine_3['Discount margin'] = combine_3['Discount profit']/combine_3['Discount sales']

# calculate discount rate
# Fomula: Discount margin = discount profit / discount sales [in which discount sale = original sale (1- discount rate)]
combine_3['Mean of discount rate'] = 1-(combine_3['Discount profit']/(combine_3['Discount margin']*combine_3['Original sales']))

# Calculate margin AFTER discount
# Fomula: org margin = dis margin*(1-discount rate)+discount rate
combine_3['Original margin'] = combine_3['Discount margin']*(1-combine_3['Mean of discount rate'])+combine_3['Mean of discount rate']

combine_3=combine_3[['Product Name','Original sales','Original margin','Discount sales','Discount profit','Discount margin','Mean of discount rate']]
combine_3.head()
```

```python
# describe dataframe
combine_3.describe()
```

```python
# plot the margin between discount and without discount
import matplotlib.pyplot as plt
plt.figure(figsize=(30, 16), tight_layout=False)
plt.scatter(combine_3['Product Name'] , combine_3['Discount margin'], color='#e3597d',label='Discount margin',s=100)
plt.scatter(combine_3['Product Name'] , combine_3['Original margin'], color='#f5cbcd',label='Original margin',s=100)
plt.axhline(y=0, color='#4d4f4c', linestyle='--')
plt.axhline(y=0.117966, color='#19bf98', linestyle='-',label='Expected margin')
plt.xticks(rotation=90, fontsize=14)
plt.yticks(fontsize=18)
plt.xlim(-1,118)
plt.legend(fontsize=18)

# identify the negative objects
plt.annotate('loss', xy=(9.5,-0.042), xytext=(11, -0.052),color='#f00543',fontsize=16,
             arrowprops=dict(color='#f00543', shrink=0.04))
plt.annotate('loss', xy=(77.5,-0.042), xytext=(79, -0.052),color='#f00543',fontsize=16,
             arrowprops=dict(color='#f00543', shrink=0.04))
plt.annotate('loss', xy=(76.5,-0.022), xytext=(78, -0.032),color='#f00543',fontsize=16,
             arrowprops=dict(color='#f00543', shrink=0.04))
plt.title('Profit margin by product with and without discount', fontweight='bold', fontsize=20)
plt.show()
```

```python
#Create a dataframe with the discount margin below the average basing on the observation
below=combine_3[combine_3['Discount margin'] < 0.118056]
below = below.reset_index(drop=True)
below.head()
```

```python
# Check the number of product having the profit margin below the average level
print('There are',below['Product Name'].nunique(), 'products having the profit margin below the average')
```

```python
# calculate the mean of NEW discount rate that the new profit margin 
# will be greater than or equal to mean value at least.

# Fomula: Discount margin = Discount profit / [Orignal sales*(1 - Discount rate)]
below['NEW mean of discount rate']= (0.118056-below['Original margin'])/(0.118056-1)

# Fomula: new margin = (old margin - discount) / (1 - discount)
below['NEW margin'] = (below['Original margin']-below['NEW mean of discount rate'])/(1-below['NEW mean of discount rate'])
below.head()
```

```python
# what product needs to be alerted in price instead of discount rate?
below_1=below[below['NEW mean of discount rate'] < 0]
below_1 = below_1.reset_index(drop=True)
below_1.head()
```

# 4. How does time reflect on sales?

```python
# set the date time data in Python format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Date'].head()
```

```python
# create a new dataframe with date vs sale
date_sale= df.groupby('Order Date',as_index=False)['Sales'].sum()
date_sale.head()
```

```python
# convert data date into numeric one because all the value has to be numeric for y prediction\
# hence, datetime data is not qualified --> convert to ordinal
import datetime as dt
date_sale['Order Date [ord]'] = date_sale['Order Date'].map(lambda x: x.toordinal())
date_sale = date_sale[['Order Date','Order Date [ord]','Sales']]
date_sale.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn import linear_model
import numpy as np

# Define the dependent and independent variable
x = date_sale['Order Date [ord]']   #independent variable
y = date_sale['Sales']   #dependent variable

z = date_sale['Order Date']
```

```python
# Exploring the dataset
plt.figure(figsize=(12, 8), tight_layout=True)
sns.scatterplot(x=z, y=y)
plt.show()
```

```python
# Making the regression: Adding a constant and fitting the model
x = sm.add_constant(x) # adding a constant
lm = sm.OLS(y,x).fit() # fitting the model
```

```python
lm.predict(x)
```

```python
# The Regression Table
lm.summary()
```

```python
# Order Date ord: -5.8088
# Constant coef: 4.309e+06

# Linear equation: 洧녽 = 洧녩洧논 + 洧녪
y_pred = -5.8088*x['Order Date [ord]'] +4.309e+06
```

```python
# Plotting The Regression Line


# plotting the data points
ax1 = plt.figure(figsize=(12, 8), tight_layout=True)
ax1 = sns.scatterplot(x=z, y=y, color='c', label='Sum of sales by mont')

#plotting the line
line=sns.lineplot(x=z,y=y_pred, color='#e3597d', label='Linear y = -5.8088x + (4.309e+06)')

#axes
plt.xlim()
plt.ylim(0,120000)
plt.legend(fontsize=14, loc='upper left')
plt.savefig('linear_regression')
plt.title('SALES BY TIME',fontweight='bold', fontsize=18)
plt.show()
```

# 5. What is the distribution of customer by RFM score?

```python
# recency

rece = df.groupby('Customer Id', as_index=False)['Order Date'].max()
rece.columns = ['Customer Id', 'Last Purchase Date']

recent_date = rece['Last Purchase Date'].max()
 
rece['Recency'] = rece['Last Purchase Date'].apply(lambda x: (recent_date - x).days)
rece.head()
```

```python
# frequency
freq=df.groupby('Customer Id', as_index=False)['Order Date'].count()
freq.columns = ['Customer Id', 'Frequency']

freq.head()
```

```python
# monetary
mone = df.groupby('Customer Id', as_index=False)['Sales'].sum()
mone.columns = ['Customer Id', 'Monetary']

mone.head()
```

```python
# Merging the columns R-F-M in a dataframe

rf = rece.merge (freq, on='Customer Id')

rfm = rf.merge (mone, on='Customer Id').drop(columns = 'Last Purchase Date')

rfm.head(10)
```

```python
# assign a ranking point for each value [ 5 is best -> 1 is worst]
rfm["R"] = pd.qcut(rfm["Recency"],q=5,labels=[5,4,3,2,1])

# Order point of recency is different the rest two because, to recency, the lower the better
```

```python
# assign a ranking point for each value [ 5 is best -> 1 is worst]
bins=[0,9,18,27,36,47]
labels = [1,2,3,4,5]
rfm['F'] = pd.cut(rfm['Frequency'], bins=bins, labels=labels)
```

```python
# assign a ranking point for each value [ 5 is best -> 1 is worst]
rfm["M"] = pd.qcut(rfm["Monetary"],q=5,labels=[1,2,3,4,5])
rfm.head(10)
```

```python
# Input the column of score in the dataframe
rfm["RFM Score"] = rfm["R"].astype(str)  +rfm["F"].astype(str)+ rfm["M"].astype(str)
rfm.head()
```

```python
# convert the data in RFM score into numeric to perform the loop to assign the customer 
# segmentation labels

rfm['RFM Score'].astype(int)
```

```python
# Attach the customer segmentation for each customer
# list of the customer segmentation basing on the RFM score
list1 = [555, 554, 544, 545, 454, 455, 445]
list2 = [543, 444, 435, 355, 354, 345, 344, 335]
list3 = [553, 551, 552, 541, 542, 533, 532, 531, 452, 451, 442, 441, 431, 453, 433, 432, 423, 353, 352, 351, 342, 341, 333, 323]
list4 = [512, 511, 422, 421 ,412, 411, 311]
list5 = [525, 524, 523, 522, 521, 515, 514, 513, 425,424, 413,414,415, 315, 314, 313]
list6 = [535, 534, 443, 434, 343, 334, 325, 324]
list7 = [155, 154, 144, 214,215,115, 114, 113]
list8 = [331, 321, 312, 221, 213]
list9 = [255, 254, 245, 244, 253, 252, 243, 242, 235, 234, 225, 224, 153, 152, 145, 143, 142, 135, 134, 133, 125, 124]
list10= [332, 322, 231, 241, 251, 233, 232, 223, 222, 132, 123, 122, 212, 211]
list11= [111, 112, 121, 131, 141, 151]

# define a function to assign the segmentation for the RFM score
def ranking(x):
    if x in list1:
        return 'Top valuable customer'
    if x in list2:
        return 'Loyal customer'
    if x in list3:
        return 'Potential loyalist'
    if x in list4:
        return 'Rookie'
    if x in list5:
        return 'Promising'
    if x in list6:
        return 'Exigent'
    if x in list7:
        return 'Cannot lose them'
    if x in list8:
        return 'About to sleep'
    if x in list9:
        return 'At risk'
    if x in list10:
        return 'Hibernating'
    if x in list11:
        return 'Slipping'
    else:
        return 'Unknown'
    
    
# mapping the value to dataframe
rfm['RFM segmentation'] = rfm['RFM Score'].astype(int).apply(ranking)  
rfm.head(10)
```

```python
# checking value of RFM segmentation
rfm['RFM segmentation'].unique()
```

```python
# Calculate the count and percentage of customers in each RFM segment
seg_counts = rfm.groupby('RFM segmentation')['RFM segmentation'].count()
seg_pct = round(seg_counts / len(rfm) * 100 , 2)

# Combine percentages into a dataframe
seg_df = pd.concat([seg_counts, seg_pct], axis=1)
seg_df.columns = ['Count', 'Percentage']

seg_df
```
